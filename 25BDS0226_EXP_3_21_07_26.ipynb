```python
import pandas as pd
import numpy as np
```

```python

d1A = pd.DataFrame({'ID': list(range(15)),
                    'temp': [15, 22, 35, 11, -2, 28, 19, 32, 25, 14, 8, 30, 21, -5, 17]})
d2A = pd.DataFrame({'ID': list(range(15)),
                    'temp': [12, 29, 31, 5, -8, 26, 18, 34, 22, 16, 3, 27, 23, -1, 13]})
d1B = pd.DataFrame({'ID': list(range(15)),
                    'rain': [12, 45, 0, 8, 110, 23, 56, 0, 15, 88, 4, 12, 67, 34, 0]})
d2B = pd.DataFrame({'ID': list(range(15)),
                    'rain': [0, 14, 78, 92, 5, 0, 43, 11, 29, 85, 0, 19, 54, 3, 72]})
```

```python

dfTemp = pd.concat([d1A, d2A], ignore_index=True)
dfRain = pd.concat([d1B, d2B], ignore_index=True)
df = pd.concat([dfRain, dfTemp], axis=1)
print("dfTemp (first 5 rows):")
display(dfTemp.head())
print("\ndfRain (first 5 rows):")
display(dfRain.head())
print("\nFinal combined df (first 5 rows):")
display(df.head())
```

***OPTION-2***

```python
dfTemp = pd.concat([d1A, d2A], ignore_index=True)
dfRain = pd.concat([d1B, d2B], ignore_index=True)
df = dfRain.merge(dfTemp, on='ID', how='inner')
print("Final merged df (first 5 rows):")
display(df.head())
```

***OPTION-3***

```python
dfTemp = pd.concat([d1A, d2A], ignore_index=True)
dfRain = pd.concat([d1B, d2B], ignore_index=True)
df = dfRain.merge(dfTemp, on='ID', how='left')
print("Final merged df (first 5 rows):")
display(df.head())
```

***OPTION-4***

```python
dfTemp = pd.concat([d1A, d2A], ignore_index=True)
dfRain = pd.concat([d1B, d2B], ignore_index=True)
df = dfRain.merge(dfTemp, on='ID', how='right')
print("Final merged df (first 5 rows):")
display(df.head())
```
